```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://webscraper.io/test-sites/e-commerce/allinone', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
print(soup.prettify())
```

```python
for i in soup.find_all('p'):
    print(i.text)
```

```python
for i in soup.find_all('p'):
    print(i.text.strip())
```
