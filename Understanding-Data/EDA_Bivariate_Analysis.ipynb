```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('EVSE-B-PowerCombined.csv')
```

```python
plt.figure(figsize=(20, 20))
sns.scatterplot(data=df, x=df['shunt_voltage'],y=df['power_mW'],hue=df['Label'])
plt.show()
```

```python
sns.boxplot(data=df, x=df['Label'],y=df['power_mW'],hue=df['Attack-Group'])
```

```python
sns.heatmap(pd.crosstab(df['Attack-Group'],df['State']))
```

```python
sns.heatmap(pd.crosstab(df['Attack-Group'],df['interface']))
```
